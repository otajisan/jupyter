```python
import json
import os
import sys
import urllib

from bs4 import BeautifulSoup
import requests

GOOGLE_IMAGE_SEARCH_URL = 'https://www.google.co.jp/search'
```

```python
session = requests.session()
session.headers.update({'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:10.0) Gecko/20100101 Firefox/10.0'})
```

```python
# Google画像検索にかけるURLを生成
keyword = '美女'
page = 0
params = urllib.parse.urlencode({
    'q': keyword,
    'tbm': 'isch',
    'sa': 'X'
})
url = GOOGLE_IMAGE_SEARCH_URL + '?' + params
url
```

```python
# スクレイピング結果をJSONに格納
html = session.get(url).text
soup = BeautifulSoup(html)
elements = soup.select('.rg_meta.notranslate')
jsons = [json.loads(e.get_text()) for e in elements]
```

```python
# 必要な要素のみ取得
image_list = []
for data in jsons:
    d = {
        'title': data['pt'],
        'short_desc': data['s'],
        'image_url':  data['ou'],
        'src_url': data['ru'],
        'cp': data['isu'],
    }
    image_list.append(d)

# 取得件数
len(image_list)
```

```python
# 取得結果
json.dumps(
    image_list,
    indent=2,
    ensure_ascii=False
)
```
